```python
data_n = data[['Position ID', 'Time']]
data_n['Time'] = data_n['Time'].dt.date
data_n = data_n.drop_duplicates(subset=['Position ID', 'Time'], keep='first')
data_n = data_n.reset_index()
data_n = data_n.drop(columns=['index'])
data_n.head()
```

```python
# Finding Employees who came consecutive 7 days

# Initialize variables
consecutive_days = 0
last_entry_date = None
consecutive_employees = []

# Iterate through the sorted data
for index,  row in data_n.iterrows():
    if last_entry_date is None or (row['Time'] - last_entry_date).days == 1:
        consecutive_days += 1
        last_entry_date = row['Time']
        # print(consecutive_days)
    else:
        consecutive_days = 1
        last_entry_date = row['Time']

    if consecutive_days == 7:
        consecutive_employees.append(row['Position ID'])

print(consecutive_employees)
```
